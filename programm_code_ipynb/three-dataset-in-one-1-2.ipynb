```python
# то есть должен быть датасет на 400 000 строк где 67% будет из movies, 
# 10.4% из office и 22% из tools
# получается из movies надо взять 68*4=272
# 4*10=40 это из office
# 4*22=88 это из tools
# получаем как раз 400 тысяч строк в датасете и пропорционально все взяли 
dataframe_movies = movies.head(272000)
dataframe_movies
```

```python
dataframe_office = office.sample(40000)
dataframe_office
```

```python
dataframe_tools = tools.sample(88000)
dataframe_tools
```

```python
# как видно все очень хорошо и пропусков нет у нас
print(dataframe_movies.isnull().sum())
print(dataframe_office.isnull().sum())
print(dataframe_tools.isnull().sum())
```

```python
# в итоге надо соединить сначала два датафрейма по pd.concat а потом этот соединить с третим
dataframe_movies_office = pd.concat([dataframe_movies, dataframe_office])
dataframe_full_without_reset_index = pd.concat([dataframe_movies_office, dataframe_tools])
dataframe_full = dataframe_full_without_reset_index.reset_index(drop=True)
dataframe_full.info()
```
