```python
import pandas as pd
import gzip

def parse(path):
    g = gzip.open(path, 'r')
    for l in g:
        name=b'"verified": \"true\",'
        l=l.replace(b'"verified": true,',bytes(name))
        name1=b'"verified": \"false\",'
        l=l.replace(b'"verified": false,',bytes(name))
        yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

office_products_full = getDF('/content/drive/MyDrive/vkr_mkdir/Office_Products_5.json.gz')
movies_and_tv_full = getDF('/content/drive/MyDrive/vkr_mkdir/Movies_and_TV_5.json.gz')
```
