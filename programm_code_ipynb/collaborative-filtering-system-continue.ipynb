Collaborative Filtering System. User-Item Matrix

```python
new_pr = pd.read_csv("/kaggle/input/collaborative-csv/data_collaborative_full_csv")
new_pr = new_pr.sample(50000)
```

```python
new_pr
```

```python
sentiment_matrix = new_pr.pivot_table(values='total_score', index='asin', columns='reviewerID', fill_value=0)
sentiment_matrix
```

```python
sentiment_matrix.shape
```

```python
X = sentiment_matrix
X.head(20)
```

Correlation Matrix. Корреляционная матрица используется в совместной фильтрации для определения сходства между элементами, которые могут быть рекомендованы пользователю. Корреляционная матрица вычисляет корреляцию между всеми парами элементов в наборе данных. Чем выше значения тем выше сходство. 

```python
#Decomposing the Matrix
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(X)
decomposed_matrix.shape
'''
The singular value decomposition(SVD) provides another way to factorize a matrix, into singular vectors and singular values. ... The SVD is used widely both in the calculation of other matrix operations, such as matrix inverse, but also as a data reduction method in machine learning
'''
decomposed_matrix.shape
```

```python
correlation_matrix = np.corrcoef(decomposed_matrix) 
correlation_matrix
```

```python
i = "0486413012"

product_names = list(X.index)
product_ID = product_names.index(i)
product_ID
```

```python
correlation_product_ID = correlation_matrix[product_ID]
correlation_product_ID.shape
```

```python
Recommend = list(X.index[correlation_product_ID > 0.65])
# Removes the item already bought by the customer
Recommend.remove(i)
Recommend[0:24]
     
```
